```python
df = df.drop('date', axis = 1)
df.columns
```

```python
df['zipcode'].value_counts()
```

```python
df = df.drop('zipcode', axis = 1)
```

```python
X = df.drop('price', axis = 1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_test.max()
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss = 'mse')
```

```python
model.fit(x = X_train, y = y_train, 
          validation_data = (X_test, y_test),
         batch_size = 128, epochs = 400
         )
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = pd.DataFrame(model.predict(X_test), columns = ['Predicted Price'])
```

```python
predictions['Y Test'] = y_test
predictions
```

```python
sns.scatterplot(x = y_test, y = predictions['Predicted Price'])
```

```python
mean_squared_error(y_test, predictions['Predicted Price'])**0.5
```

```python
mean_absolute_error(y_test, predictions['Predicted Price'])
```

```python
explained_variance_score(y_test, predictions['Predicted Price'])
```
